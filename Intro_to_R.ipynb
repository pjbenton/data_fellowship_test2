### **Control Flows in R**

```r
# control flows means using logic statements ie(=, !=, <, <=, etc) these are the same in R as in python
```

```r
# python syntax example

value = 10
if value > 5:
    print('something')
```

```r
value <- 10
if (value > 5){
    print('greater than 5')
}
```

```r
value <- 4
if (value > 5){
    print('greater than 5')
} else {
    print('less than or equal to 5')
}
```

```r
value <- 11

if (value > 10){
    print('greater than 10')
} else if (value == 5){
    print('equals 5')
} else {
    print('not 5 and less than or equal to 10')
}
```

```r
value <- 7

if (value > 5 & value < 10){
    print('value is between 5 and 10')
}
```

```r
value <- 2

if (value > 5 | value == 2){
    print('value is 2 or greater than 5')
}
```

```r
# & = AND
# | = OR
```

### **T-test**

```r
# check whether a correlation is significant - whether this is by chance or not
```

```r
gap %>% filter(year == 1977) %>% group_by(continent) %>% summarise(mean(life_exp))
```

```r
gap %>% filter(year == 2007) %>% group_by(continent) %>% summarise(mean(life_exp))
```

```r
# is this increase significant?
```

```r
gap_asia_2007 <- gap %>% filter(year == 2007, continent == 'Asia')
```

# H0: The increase in life expectancy is not significant

# H1: The increase in life expectancy IS significant

T-test - 5%

```r
t.test(gap_asia_2007$life_exp, mu=59.6)

# comparing the means of life expectancy - 59.6 is life_exp of Asia in 1977
```

```r
# p-value = 0.00000000364

# because of p value is < 0.05

# we can accept the alternative hypothesis = H1

# statistically speaking life expectancy was no the same / this change was not due to chance
```
