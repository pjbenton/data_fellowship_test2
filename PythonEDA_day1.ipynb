```python
import pandas as pd
```

```python
data = {
    'apples': [2, 4, 6, 8],
    'oranges': [1, 3, 5, 7]
}

purchases = pd.DataFrame(data)

purchases
```

```python
titanic = pd.read_csv('titanic.csv')

titanic
```

```python
#slice dataframe selecting a single column

titanic['survived']

titanic.survived

titanic.loc[:,'survived']

#^^^multiple syntax which deliver same result

##Pandas module enables us to handle large data sets containing a considerably huge amount of data for processing altogether.
##This is when Python loc() function comes into the picture. 
##The loc() function helps us to retrieve data values from a dataset at an ease.
##Using the loc() function, we can access the data values fitted in the particular row or column based on the index value passed to the function.
```

```python
#slice dataframe selecting a multiple columns

titanic[['survived', 'pclass']]

titanic.loc[:, ['survived', 'pclass']]

#^^multiple syntax which deliver same result
```

```python
#slice a dataframe selecting a single row

titanic.loc[0]
```

```python
#slice a dataframe selecting a multiple row

titanic.loc[[0, 1]]
```

```python
#slice a range

titanic.loc[range(0,5)]
```

```python
#slice specific datapoints

titanic.loc[0, 'name']
```

```python
#Filter and Sort

titanic[titanic['sex'] == 'female']
```

```python
titanic[titanic['fare'] >= 50.5].head(5)
```

```python
#several conditions

titanic[
    ((titanic['age'] >= 45) & (titanic['age'] <= 65))
    & (titanic['fare'] > 100.000)
    & (titanic['embarked'] == 'C')
]
```

```python
titanic.sort_values('fare')
```

```python
titanic.sort_values('fare', ascending = False)
```

```python
titanic.filter(like='s', axis = 1)

#filter just looks at headers for this syntax
```

```python
#can use this syntax for checking specific datasets

titanic.loc[titanic['name'].str.contains("mabel", case=False)]
```

```python
#missing values

titanic.isnull()
```

```python
titanic.isnull().sum()
```

```python
#split-apply-combine

titanic.groupby('sex').mean()
```

```python
titanic.groupby('survived').mean()
```

```python
import numpy as np
```

```python
#square root of purchases

purchases.apply(np.sqrt)
```

```python
#axis 0 = row

purchases.apply(np.sum, axis = 0)
```

```python
#axis 1 = column

purchases.apply(np.sum, axis = 1)
```

```python
def calculate_area(row):
    return row['apples'] * row['oranges']

calculate_area
```

```python
purchases.apply(calculate_area, axis = 1)
```
